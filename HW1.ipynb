```python
import numpy as np


# I
def f(x):
    return x * np.sin(3 * x) - np.exp(x)

def df(x):
    return np.sin(3 * x) + 3 * x * np.cos(3 * x) - np.exp(x)

def newton(x0, tol=1e-6):
    x_values = []
    x = x0
    for i in range(100):
        x_values.append(x)
        x_new = x - f(x) / df(x)
        if abs(f(x)) < tol:
            x_values.append(x_new)
            break
        x = x_new
    return np.array(x_values), i + 1

x1 = -0.7
x2 = -0.4
def bisection(x1, x2):
    x_mid = []
    for i in range(100):
        mid = (x1 + x2) / 2
        x_mid.append(mid)
        if f(mid) > 0:
            x1 = mid
        else:
            x2 = mid
        if abs(f(mid)) < 1e-6:
            break
    return np.array(x_mid), i + 1

A1, iter_newton = newton(-1.6)
A2, iter_bisection = bisection(-0.7, -0.4)
A3 = np.array([iter_newton, iter_bisection])

print("A1 =", A1)
print("A2 =", A2)
print("A3 =", A3)

# II
A = np.array([[1, 2], [-1, 1]])
B = np.array([[2, 0], [0, 2]])
C = np.array([[2, 0, -3], [0, 0, -1]])
D = np.array([[1, 2], [2, 3], [-1, 0]])
x = np.array([1, 0])
y = np.array([0, 1])
z = np.array([1, 2, -1])

A4 = A + B
A5 = 3 * x - 4 * y
A6 = A @ x
A7 = B @ (x - y)
A8 = D @ x
A9 = D @ y + z
A10 = A @ B
A11 = B @ C
A12 = C @ D  

print("A4 =", A4)
print("A5 =", A5)
print("A6 =", A6)
print("A7 =", A7)
print("A8 =", A8)
print("A9 =", A9)
print("A10 =", A10)
print("A11 =", A11)
print("A12 =", A12)
```
